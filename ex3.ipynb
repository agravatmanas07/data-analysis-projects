```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_table(r'/content/campaigns_delivery.tsv', sep='\t')
data.describe()

```

```python
data.head()
```

```python
data['Margin'] = data['Criteo Revenue'] - data['Criteo Cost']
data.describe()
```

```python
data.info()
```

```python
plt.figure(figsize=(10, 6))
avg_ov_per_0s=data.groupby('Os')['Margin'].mean()
avg_ov_per_0s.plot(kind='bar')
plt.ylabel('Margin')
plt.xlabel('Os')
plt.title('Average Order Value per Operating System')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
avg_ov_per_0s=data.groupby('Environment')['Margin'].mean()
avg_ov_per_0s.plot(kind='bar')
plt.ylabel('Margin')
plt.xlabel('Environment')
plt.title('Average Order Value per Operating System')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

data['Day'] = pd.to_datetime(data['Day'])

average_margin_per_day = data.groupby('Day')['Margin'].mean()


plt.figure(figsize=(10, 6))
plt.plot(average_margin_per_day.index, average_margin_per_day.values)
plt.xlabel('Day')
plt.ylabel('Average Margin')
plt.title('Average Margin Over Time')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Sample dataset
data = {'x': [1, 2, 3, 4, 5], 'y': [2, 4, 5, 4, 5]}
df = pd.DataFrame(data)

# Splitting data into x and y
X = df[['x']]  # Independent variable
y = df['y']    # Dependent variable

# Model instantiation and fitting
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Plotting the regression line
plt.scatter(X['x'], y, color='blue')
plt.plot(X['x'], y_pred, color='red', label='Fitted Line')
plt.title('Simple Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Evaluating the model
mse = mean_squared_error(y, y_pred)
print(f'Mean Squared Error: {mse}')
```
